```python
%%writefile app.py
import streamlit as st
import joblib
model=joblib.load('knn_model')
st.title("Fruits Prediction")
st.subheader("This app is used to predict fruits based on mass,width and heigth. ")
st.subheader("Fruits are categorized as apple,lemon,mandarin and orange.")
st.text("Provide input as mass,width and heigth and click on predict to get the desire fruit name.")

m=st.number_input("Enter the mass",min_value=50,max_value=400,value=50,step=1)
w=st.number_input("Enter the width",min_value=1.0,max_value=10.0,value=1.0,step=0.1)
h=st.number_input("Enter the heigth",min_value=1.0,max_value=10.0,value=1.0,step=0.1)
op=model.predict([[m,w,h]])

if st.button("PREDICT"):
   st.header("According to given mass"+str(m)+" kg , width"+str(w)+" cm and heigth"+str(h)+" cm the predicted fruits is "+str(op[0]))
```
