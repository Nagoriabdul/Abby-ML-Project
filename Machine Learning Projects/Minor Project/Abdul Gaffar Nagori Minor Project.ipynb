```python
X=da.iloc[:,[0,1,2]].values
X
```

```python
y=da.iloc[:,3].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=3,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=13,metric='euclidean')
```

```python
model.fit(X_train_scaled,y_train)
```

```python
model.predict([[3.5,2.6,1.5]])
```

```python
y_p=model.predict(X_test_scaled)
y_p
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,precision_score,recall_score
```

```python
accuracy_score(y_p,y_test)
```

```python
print(classification_report(y_p,y_test))
```

```python
pd.DataFrame(confusion_matrix(y_p,y_test),index=['setosa_A','versicolor_A','virginica_A'],columns=['setosa_P','versicolor_P','virginica_P'])
```

```python
%%writefile app.py
import streamlit as st
import joblib
model=joblib.load('knn_model')
st.title("Iris Flower Species Prediction")
st.subheader("This app classify species based on Sepal Width,Length and Petal width. ")
st.text("Provide input as Sepal width,Length and Petal width and click on Predict to get Species name!!")

sw=st.number_input("Enter the Sepal Width",min_value=2.0,max_value=5.0,value=2.0,step=0.1)
sl=st.number_input("Enter the Sepal Length",min_value=4.0,max_value=8.0,value=4.0,step=0.1)
pw=st.number_input("Enter the Petal Width",min_value=0.1,max_value=2.50,value=0.1,step=0.1)
op=model.predict([[sw,sl,pw]])

if st.button("PREDICT"):
   st.header("According to your Sepal width"+str(sw)+", Sepal length"+str(sl)+"and Petal width"+str(pw)+"your desire "+str(op[0])+" species.")
```
